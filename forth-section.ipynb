####  Seaborn

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = 8 , 4
import seaborn as sns
%matplotlib inline
```

```python
vis1 = sns.displot(stats['InternetUsers'], kde=True) ## For discription and additional information press shift+tab
```

```python
vis1 = sns.displot(stats['InternetUsers'], kde=True, bins=30)
```

```python
vis2 = sns.boxplot(data=stats, x='IncomeGroup', y='BirthRate')
```

Seaborn has many types of plots, vist their website for more.
Seaborn is built on matplotlib.

```python
## linear regresion plot
## hue, paramter that paint the points on the plots based on a specific column values
vis3 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate', hue='IncomeGroup', height=10, aspect=1) 
```

kws - keyboard arguments <br>
Seaborn is using matplotlib behind the scene. If we want to pass parameters to seaborn to be used in matplotlib we can pass dictionaries that will be passed to matplotlib. For example, scatter of plot.

```python
vis3 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate', hue='IncomeGroup', height=10, aspect=1, scatter_kws={'s': 100}) ## S is a scalar for the size of points n^2
```
